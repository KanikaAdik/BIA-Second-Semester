```python
df = df.fillna(0)
```

```python
#Charts n graphs

#changing columns with "f" & "t" to binary variables

binary_columns = ["host_is_superhost","host_has_profile_pic","host_identity_verified", "instant_bookable"]
for col in binary_columns:
    df.loc[(df[col] == 'f' ), col] = 0
    df.loc[(df[col] == 't' ), col] = 1
    
df['host_has_profile_pic'].astype(int)
df['host_identity_verified'].astype(int)
df['instant_bookable'].astype(int)

```

```python
#changing room type to 1 2 3 4 each for Entire home/apt, Private room, Shared room, hotel room

room_types =['Entire home/apt', 'Private room', 'Shared room','Hotel room' ]
for value in room_types:
    if value =='Entire home/apt':
        df.loc[(df['room_type'] == value ), 'room_type'] = 1
    if value =='Private room':
        df.loc[(df['room_type'] == value ), 'room_type'] = 2
    if value =='Shared room':
        df.loc[(df['room_type'] == value ), 'room_type'] = 3
    if value =='Hotel room':
        df.loc[(df['room_type'] == value ), 'room_type'] = 4

df = df['room_type'].astype(int)
df.head()
```

```python
df['host_is_superhost']=df['host_is_superhost'].astype(int)
```

```python
#Changing Price to float values
df['price'] = df['price'].replace('[\$\,]', '', regex=True).astype(float)
df.apply(lambda col: len(col.unique()))
```
